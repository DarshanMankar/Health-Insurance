```python
def impute_nan(test_df,variable,median):
    test_df[variable+"_median"]=test_df[variable].fillna(median)
```

```python
median=test_df.Holding_Policy_Type.median()
median
```

```python
impute_nan(test_df,'Holding_Policy_Type',median)
test_df.head()
```

```python
print(test_df['Holding_Policy_Type'].std())
print(test_df['Holding_Policy_Type_median'].std())
```

```python
test_df=test_df.drop(['Holding_Policy_Duration'],axis=1)
test_df['Holding_Policy_Duration_median']=test_df['Holding_Policy_Duration_median'].astype(float)
```

```python
test_df["Health Indicator"].value_counts()
```

```python
def impute_nan_object(test_df,variable):
    test_df[variable+" new var"]=np.where(test_df[variable].isnull(),"Missing",test_df[variable])
```

```python
for feature in ['Health Indicator']:
    impute_nan_object(test_df,feature)
```

```python
test_df.head()
```

```python
test_df=test_df.drop(['Health Indicator'],axis=1)
test_df=test_df.drop(['ID'],axis=1)
```

```python
test_df.head()
```

```python
Accomodation=pd.get_dummies(test_df['Accomodation_Type'],drop_first=True)
Reco_Insurance=pd.get_dummies(test_df['Reco_Insurance_Type'],drop_first=True)
Spouse=pd.get_dummies(test_df['Is_Spouse'],drop_first=True)
Health_Indicator=pd.get_dummies(test_df['Health Indicator new var'],drop_first=True)
```

```python
test_df = pd.concat([test_df,Accomodation,Reco_Insurance,Spouse,Health_Indicator],axis=1)
```

```python
test_df.drop(['Accomodation_Type','Reco_Insurance_Type','Is_Spouse','Health Indicator new var'],axis=1,inplace=True)
test_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
test_df["City_Code"]=encoder.fit_transform(test_df['City_Code'])
```
