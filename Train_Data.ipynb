```python
def impute_nan(train_df,variable,median):
    train_df[variable+"_median"]=train_df[variable].fillna(median)
```

```python
impute_nan(train_df,'Holding_Policy_Type',median)
train_df.head()
```

```python
def impute_nan_object(train_df,variable):
    train_df[variable+" new var"]=np.where(train_df[variable].isnull(),"Missing",train_df[variable])
```

```python
for feature in ['Health Indicator']:
    impute_nan_object(train_df,feature)
```

```python
train_df=train_df.drop(['Health Indicator'],axis=1)
train_df=train_df.drop(['ID'],axis=1)
```

```python
Accomodation=pd.get_dummies(train_df['Accomodation_Type'],drop_first=True)
Reco_Insurance=pd.get_dummies(train_df['Reco_Insurance_Type'],drop_first=True)
Spouse=pd.get_dummies(train_df['Is_Spouse'],drop_first=True)
Health_Indicator=pd.get_dummies(train_df['Health Indicator new var'],drop_first=True)
```

```python
train_df = pd.concat([train_df,Accomodation,Reco_Insurance,Spouse,Health_Indicator],axis=1)
```

```python
train_df.drop(['Accomodation_Type','Reco_Insurance_Type','Is_Spouse','Health Indicator new var'],axis=1,inplace=True)
train_df.head()
```
